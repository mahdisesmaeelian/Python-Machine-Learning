```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
N = 150
std = 0.6

melons = pd.DataFrame({
    'width': np.random.normal(8, std, N),
    'height': np.random.normal(5, std, N),
    'weight': np.random.normal(8, std, N)})

ballons = pd.DataFrame({
    'width': np.random.normal(12, std, N),
    'height': np.random.normal(24, std, N),
    'weight': np.random.normal(1, std, N)})
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.scatter3D(melons['width'],melons['height'],melons['weight'], color = "orange")
ax.scatter3D(ballons['width'],ballons['height'],ballons['weight'], color = "red")

ax.set_xlabel('width')
ax.set_ylabel('height')
ax.set_zlabel('weight')

plt.title("Details")
plt.show()
```
