```python
N = 500
apple_width = np.random.normal(6,2,N)
apple_length = np.random.normal(6,2,N)

banana_width = np.random.normal(4,2,N)
banana_length = np.random.normal(8,2,N)
```

```python
plt.scatter(apple_width,apple_length , c ="red" , label = "apple")
plt.scatter(banana_width,banana_length , c = "yellow" , label = "banana")
plt.xlabel("Width")
plt.ylabel("Length")
plt.legend()
```
