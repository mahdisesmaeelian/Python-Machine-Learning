```python
import numpy as np
from numpy.linalg import inv
import matplotlib.pyplot as plt
from scipy import stats
```

```python
X = np.random.uniform(0,15,200)
Y = X+ np.random.uniform(0,5,200)
slope, intercept, r_value, p_value, std_err = stats.linregress(X,Y)

plt.scatter(X ,Y)
```

```python
Y_pred = X * np.random.uniform(0,2)

plt.scatter(X ,Y,label='original data')
plt.plot(X ,Y_pred, c= 'red', lw = 4,label='fitted line')
plt.legend()
```

```python
X = X.reshape((200,1))
Y = Y.reshape((200,1))
```

```python
m = np.matmul(inv(np.matmul(X.T,X)), np.matmul(X.T,Y))
m
```

```python
Y_pred = np.matmul(X,m)
```

```python
plt.scatter(X ,Y)
plt.plot(X ,Y_pred, c= 'red', lw = 4)
```

```python
plt.plot(X ,Y, 'o', label='original data')
plt.plot(X ,Y_pred, c= 'red', label='fitted linear least squares')
plt.plot(X, intercept + slope * X, 'hotpink', label='fitted linregress line')

plt.legend()
plt.show()
```
