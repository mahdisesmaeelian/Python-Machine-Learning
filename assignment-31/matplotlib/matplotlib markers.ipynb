```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = '*')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o--m')
plt.show()
```

```python
# marker size(ms) and color(mec) and face color(mfc):
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints,marker = 'o', ms = 15 , mec = 'r',mfc = 'm')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = '#4CAF50', mfc = 'hotpink')
plt.show()
```
